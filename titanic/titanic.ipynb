```python

categories = np.arange(8)

# Get counts for each category, ensuring all values 0-7 are present
sibsp_counts = df_train['SibSp'].value_counts().reindex(categories, fill_value=0).sort_index()
parch_counts = df_train['Parch'].value_counts().reindex(categories, fill_value=0).sort_index()

# Number of categories
N = len(categories)

# Compute the angle for each category
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

# Set bar width as a fraction of the available space
bar_width = (2 * np.pi / N) * 0.4

# Create a polar subplot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

# Define an offset so the two bars for each category don't overlap
offset = bar_width / 2

# Plot bars for SibSp (shifted slightly to the left)
bars_sibsp = ax.bar(angles - offset, sibsp_counts, width=bar_width,
                    color='orange', alpha=0.6, label='SibSp')

# Plot bars for Parch (shifted slightly to the right)
bars_parch = ax.bar(angles + offset, parch_counts, width=bar_width,
                    color='red', alpha=0.6, label='Parch')

# Set the category labels at the corresponding angles
ax.set_xticks(angles)
ax.set_xticklabels(categories)

# Optional: add a legend
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

### Model Creation 

#### Problem Domain: `Discrete Binomial`
#### Possible models: `Logistic Regression` , `Decision Tree` , `Random Forest

```python
df_train.drop(df_train[["PassengerId"]],inplace=True,axis=1)
test_id = df_test['PassengerId']
df_test.drop(df_test[["PassengerId"]],inplace=True,axis=1)
```

```python
X_train=df_train.drop('Survied',axis=1)
Y_train=df_train["Survied"]

X_test=df_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)
```

```python
y_pre=rf_model.predict(X_test)
```

```python
y_pre
```

```python
from sklearn.model_selection import cross_val_score


# Perform cross-validation on the training set
cv_scores = cross_val_score(rf_model, X_train, Y_train, cv=5, scoring='accuracy')

# Print the cross-validation scores
print(f"Cross-Validation Accuracy Scores: {cv_scores}")
print(f"Mean Cross-Validation Accuracy: {cv_scores.mean():.4f}")
```

```python
# Create submission DataFrame
submission = pd.DataFrame({
    'PassengerId': test_id,  # Use the saved test IDs
    'Survived': y_pre  # Predicted SalePrice
})

# Save to CSV
submission.to_csv('submission01.csv', index=False)
```
